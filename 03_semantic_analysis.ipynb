```python
import pandas as pd
import re
import spacy

from spacytextblob.spacytextblob import SpacyTextBlob
from spacymoji import  Emoji

display.max_colwidth = 100

RANDOM_STATE = 42

nlp = spacy.load('en_core_web_md')
```

```python
nlp.add_pipe('emoji', first=True)
nlp.add_pipe('spacytextblob')
```

```python
df_full = pd.read_csv('data/tweets_users_aug_21.csv', parse_dates=['date'])

keep = ['url_x', 'date', 'content', 'renderedContent', 'id_x', 
        'user','outlinks', 'username', 'displayname', 'location',
       'followersCount', 'friendsCount']

df = df_full.loc[:, keep]
```

```python
df.sample(25, random_state=RANDOM_STATE)
```

```python
test = df.at[194, "content"]

test
```

```python
doc = nlp(test)
```

```python
doc._.polarity
```

```python
doc._.assessments
```

```python
a = [d.text for d in doc if not d.text.startswith('@')]
a_2 = " ".join(a)
a_2
```

```python
doc2 = nlp(a_2)
print(f'polarity: {doc2._.polarity}')
```

```python
# Retrieve the default token-matching regex pattern
re_token_match = spacy.tokenizer._get_regex_pattern(nlp.Defaults.token_match)

# Add #hashtag pattern
re_token_match = f"({re_token_match}|#\\w+)"

nlp.tokenizer.token_match = re.compile(re_token_match).match

```

```python
doc2 = nlp(test)
doc2
```

```python
b = [d.text for d in doc2 if not d.text.startswith('@')]
b = " ".join(b)
b
```

```python
for i in nlp(b):
    if i._.emoji_desc:
        print(i._.emoji_desc)
```
